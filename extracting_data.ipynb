```python
# Imports
import zipfile
import wget
import ftplib
import requests
from io import BytesIO
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
import os
warnings.filterwarnings('ignore')
```

```python
url = 'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/ITR/DADOS/itr_cia_aberta_2022.zip'
wget.download(url, 'itr_cia_aberta_2022.zip')
```

```python
r = requests.get(url)
z = zipfile.ZipFile(BytesIO(r.content))
files = z.namelist()
files
```

```python
dre = pd.read_csv(z.open(files[-4]), sep=';', encoding='latin-1')
dre
```
